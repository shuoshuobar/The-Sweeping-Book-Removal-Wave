```python
%%R -w 800 -h 500

# Parse the banned reasons: split each string by comma and trim whitespace.
all_reasons <- unlist(strsplit(reasons_list, ","))
all_reasons <- trimws(all_reasons)

# Create a data frame for counting
df_reasons <- data.frame(Reason = all_reasons)

# Count each banned reason
reason_counts <- as.data.frame(table(df_reasons$Reason))

# Prepare labels and sizes for the treemap
labels <- paste(reason_counts$Var1, "\n(", reason_counts$Freq, ")", sep="")
sizes <- reason_counts$Freq
```

```python
%%R -w 860 -h 600

# Define custom colors
customized_colors_2 <- c(
  "#A69E91", # Sandstone Beige
  "#5C6A6B", # Slate Gray
  "#A85E59", # Warm Coral
  "#B18F88", # Soft Cinnamon
  "#6A7F7D", # Dusty Teal
  "#537D6D", # Forest Green
  "#9A8D99", # Pale Lavender
  "#839B7A", # Olive Drab
  "#BFA59B", # Taupe Brown
  "#6E6B7E"  # Slate Blue
)

# Create a proportional font size scale
min_font_size <- 20
max_font_size <- 30 

# Normalize sizes to create proportional font sizes
max_size <- max(sizes)
min_size <- min(sizes)
scaled_font_sizes <- (min_font_size + (max_font_size - min_font_size) * (sizes - min_size) / (max_size - min_size))

# Create a treemap plot using ggplot2 and treemapify
treemap <- ggplot(reason_counts) +
  aes(area = sizes, fill = Var1, label = labels, subgroup = Var1) + 
  geom_treemap(color = "white", size = 0.5, radius = unit(5, "pt")) + 
  geom_treemap_subgroup_text(
    place = "topleft",  
    color = "white",       
    family = "Georgia", 
    fontface = "bold",
    size = scaled_font_sizes  # Adjust the label size based on the frequency/size
  ) + 
  scale_fill_manual(values = customized_colors_2) + 
  labs(
    title = "Many more books removed from shelves over sexual content and abuse themes",
    # subtitle = "Subject matter of top ten banned book titles from 2021 to 2024",
    caption = "Data source: PEN America | Treemap: Yushuo Wang",
    fill = "Reason"
  ) + 
  theme_minimal() +
  theme(
    legend.position = "none",
    text = element_text(family = "Georgia"),
    plot.title = element_text(size = 22, hjust = 0, color = "black", face = "bold", family = "Arial"),
    plot.subtitle = element_text(size = 18, hjust = 0, color = "black", family = "Arial"),
    plot.caption = element_text(size = 18, hjust = 0, color = "black", family = "Arial"),
    plot.background = element_rect(fill = "white", color = NA), 
    panel.background = element_rect(fill = "white", color = NA)  
  )

# Show the treemap
treemap
```

```python
import matplotlib.pyplot as plt
import squarify  

# Data for the top-10 banned books for three time periods (total 30 entries)
# For each entry, list the banned reasons as given:
reasons_list = [
    # 2023-2024
    "Sexual Content/Abuse, LGBTQ+ Content, Violence/Racism Concerns",
    "Sexual Content/Abuse, Substance Use/Abuse, Mental Illness",
    "Sexual Content/Abuse, LGBTQ+ Content, Substance Use/Abuse, Mental Illness",
    "Sexual Content/Abuse, Violence/Racism Concerns",
    "Substance Use/Abuse, Sexual Content/Abuse",
    "Sexual Content/Abuse, Mental Illness, Substance Use/Abuse",
    "Sexual Content/Abuse, Mental Illness, Substance Use/Abuse",
    "Sexual Content/Abuse, Violence/Racism Concerns",
    "Sexual Content/Abuse, Violence/Racism Concerns",
    "Sexual Content/Abuse",
    # 2022-2023
    "Sexual Content/Abuse, Substance Use/Abuse, LGBTQ+ Content",
    "Sexual Content/Abuse, Violence/Racism Concerns",
    "Sexual Content/Abuse",
    "Sexual Content/Abuse, Substance Use/Abuse, Mental Illness",
    "Sexual Content/Abuse, LGBTQ+ Content",
    "Sexual Content/Abuse, LGBTQ+ Content, Substance Use/Abuse, Mental Illness",
    "Sexual Content/Abuse, Violence/Racism Concerns",
    "Sexual Content/Abuse, Mental Illness, Substance Use/Abuse",
    "Substance Use/Abuse, Sexual Content/Abuse",
    "Sexual Content/Abuse",
    # 2021-2022
    "Sexual Content/Abuse, LGBTQ+ Content",
    "Sexual Content/Abuse, LGBTQ+ Content",
    "Sexual Content/Abuse, Violence/Racism Concerns",
    "Sexual Content/Abuse, Violence/Racism Concerns",
    "Sexual Content/Abuse, LGBTQ+ Content",
    "Violence/Racism Concerns",
    "Sexual Content/Abuse, Violence/Racism Concerns",
    "Sexual Content/Abuse",
    "Sexual Content/Abuse, Mental Illness, Substance Use/Abuse",
    "Sexual Content/Abuse, Violence/Racism Concerns"
]

# Parse the banned reasons: split each string by comma and trim whitespace.
all_reasons = []
for reason_str in reasons_list:
    # Split by comma and strip whitespace from each sub-reason
    parts = [r.strip() for r in reason_str.split(',')]
    all_reasons.extend(parts)

# Create a DataFrame for counting
df_reasons = pd.DataFrame(all_reasons, columns=["Reason"]) # "Reason" as column name

# Count each unique reason
reason_counts = df_reasons["Reason"].value_counts()

# Prepare labels and sizes for the treemap
labels = [f"{k}\n({v})" for k, v in reason_counts.items()] # .items() returns pairs of (key, value)
sizes = reason_counts.values # .values gives the 

# Create a treemap plot using squarify
plt.figure(figsize=(12, 8))
# Plot the treemap
ax = squarify.plot(sizes=sizes, label=labels, alpha=.8, color=plt.cm.Paired(range(len(sizes))))

# Define a font size scale
min_font_size = 12
max_font_size = 18 

# Normalize sizes to create proportional font sizes
max_size = max(sizes)
min_size = min(sizes)
scaled_font_sizes = [(min_font_size + (max_font_size - min_font_size) * (size - min_size) / (max_size - min_size)) for size in sizes]

# Adjust the font size of the labels based on the rectangle size
for i, text in enumerate(ax.texts):  # Iterate over label text objects
    text.set_fontsize(scaled_font_sizes[i])  # Set the scaled font size for each label


# Set the title and subtitle on the same line, aligned to the left
plt.title("Many more books removed from shelves over sexual content and abuse themes", fontsize=19, loc='left',fontweight='light', fontname='Arial')

# Adjust layout to reduce space between title/subtitle and plot
plt.subplots_adjust(top=0.9, bottom=0.07)  # Adjust 'top' to reduce space above the plot

# Add the data source as a caption, left-aligned and correctly spaced
plt.figtext(0.124, 0.02, "Subject matter of top ten banned book titles from 2021 to 2024\nData source: PEN America, by Yushuo Wang", ha="left", fontsize=12, fontname='Arial')

plt.axis('off')
plt.show()
```
